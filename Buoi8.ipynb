```python
class Cylinder:
    def __init__(self, r, h):
        self._r = r
        self._h = h
    def surface_area(self):
        return f"{3.14 * 2 * self._r * self._h + 2 * 3.14 * self._r * self._r:.2f}"
    def volume(self):
        return f"{3.14 * self._r * self._r * self._h:.2f}"
r, h = map(float, input().split())
cylinder = Cylinder(r, h)
print(cylinder.surface_area(), cylinder.volume())
```

```python
class Date:
    def __init__(self, dob):
        self._dob = dob
    def nhuan(self):
        y = int(self._dob[6:10])
        return (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0)
    def check_dob(self):
        if len(self._dob) != 10:
            return False
        if self._dob[2] != '/' or self._dob[5] != '/':
            return False
        d = int(self._dob[0:2])
        m = int(self._dob[3:5])
        if m < 1 or m > 12:
            return False
        if m == 2:
            if self.nhuan() == True:
                if d < 1 or d > 29:
                    return False
            else:
                if d < 1 or d > 28:
                    return False
        elif m in {4, 6, 9, 11}:
            if d < 1 or d > 30:
                return False
        else:
            if d < 1 or d > 31:
                return False
        return True
    def next_day(self):
        d = int(self._dob[0:2])
        m = int(self._dob[3:5])
        y = int(self._dob[6:10])
        if m == 2:
            maxd = 29 if self.nhuan() == True else 28
        elif m in [4, 6, 9, 11]:
            maxd = 30
        else:
            maxd = 31
        d += 1
        if d > maxd:
            d = 1
            m += 1
            if m > 12:
                m = 1
                y += 1
        self._dob = f"{d:02d}/{m:02d}/{y}"
        return self._dob
dob = input()
date = Date(dob)
if date.check_dob() == False:
    print("INVALID")
else:
    print(date.next_day())
```

```python
from math import sqrt
class Point:
    def __init__(self, x, y):
        self._x = x
        self._y = y
    def check_point(self):
        if ((self._x == 0 or self._y == 0)):
            return True
        return False
    def distance(self):
        dis = sqrt((self._x) ** 2 + (self._y) ** 2)
        return f"{dis:.2f}"
x, y = map(float, input().split())
point = Point(x, y)
print(point.check_point())
print(point.distance())
```

```python
list_of_product = [
    {"bút chì":5000},
    {"bút bi":5000}
]
class Shopping_Cart:
    def __init__(self):
        self._products = []
    def add(self, choice):
        if (choice == "1"):
            self._products.append(list_of_product[0])
        elif (choice == "2"):
            self._products.append(list_of_product[1])
    def Remove(self, choice):
        if (choice == "1"):
            if ({"bút chì":5000} in self._products):
                self._products.remove({"bút chì":5000})
        elif (choice == "2"):
            if ({"bút bi":5000} in self._products):
                self._products.remove({"bút bi":5000})
    def check_empty(self):
        if (len(self._products) == 0):
            print("True")
        else:
            print("False")
    def total(self):
        Sum = 0
        for i in self._products:
            Sum += next(iter(i.values()))
        print(Sum)
    def display(self):
        for i in self._products:
            for k, v in i.items():
                print(k, ":", v)
    def Remove_all(self):
        self._products = []
cart = Shopping_Cart()
while True:
    print("0. Thoát")
    print("1. Thêm vào giỏ hàng")
    print("2. Xóa khỏi giỏ hàng")
    print("3. Kiểm tra giỏ hàng có rỗng không")
    print("4. Tính tổng tiền giỏ hàng")
    print("5. Hiện thị toàn bộ giỏ hàng")
    print("6. Xóa toàn bộ giỏ hàng")
    choice = input()
    if (choice == "0"):
        break
    elif (choice == "1"):
        print()
        print("1. Bút chì")
        print("2. Bút bi")
        choice_product = input()
        cart.add(choice_product)
    elif (choice == "2"):
        print()
        print("1. Bút chì")
        print("2. Bút bi")
        choice_product = input()
        cart.Remove(choice_product)
    elif (choice == "3"):
        cart.check_empty()
    elif (choice == "4"):
        cart.total()
    elif (choice == "5"):
        cart.display()
    elif (choice == "6"):
        cart.Remove_all()
    print()
```

```python
class Rectangle:
    def __init__(self, w, h):
        self._w = w
        self._h = h
    def area(self):
        return self._w * self._h
    def perimeter(self):
        return 2 * (self._w + self._h)
    def scale(self, k):
        self._w = self._w * k
        self._h = self._h * k
w, h, k = map(float, input().split())
rect = Rectangle(w, h)
rect.scale(k)
print(f"{rect.area():.2f} {rect.perimeter():.2f}")
```

```python
class Fraction:
    def __init__(self, num, den):
        if den < 0:
            num = -num
            den = -den
        a = abs(num)
        b = abs(den)
        while b != 0:
            a, b = b, a % b
        self._num = num // a
        self._den = den // a
    def add(self, other):
        num = self._num * other._den + other._num * self._den
        den = self._den * other._den
        return Fraction(num, den)
    def sub(self, other):
        num = self._num * other._den - other._num * self._den
        den = self._den * other._den
        return Fraction(num, den)
    def mul(self, other):
        num = self._num * other._num
        den = self._den * other._den
        return Fraction(num, den)
    def div(self, other):
        num = self._num * other._den
        den = self._den * other._num
        return Fraction(num, den)
    def __str__(self):
        return str(self._num) + "/" + str(self._den)
a, b, op, c, d = input().split()
a = int(a)
b = int(b)
c = int(c)
d = int(d)
f1 = Fraction(a, b)
f2 = Fraction(c, d)
if op == "+":
    res = f1.add(f2)
elif op == "-":
    res = f1.sub(f2)
elif op == "*":
    res = f1.mul(f2)
elif op == "/":
    res = f1.div(f2)
print(res)
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self._owner = owner
        self._balance = balance
    def deposit(self, x):
        self._balance = self._balance + x
    def withdraw(self, x):
        if x <= self._balance:
            self._balance = self._balance - x
    def __str__(self):
        return self._owner + ": " + f"{self._balance:.2f}"
owner, balance = input().split()
balance = float(balance)
account = BankAccount(owner, balance)
n = int(input())
for i in range(n):
    cmd, x = input().split()
    x = float(x)
    if cmd == "DEPOSIT":
        account.deposit(x)
    elif cmd == "WITHDRAW":
        account.withdraw(x)
print(f"{account._balance:.2f}")
```

```python
class Student:
    def __init__(self, name):
        self._name = name
        self._scores = []
    def add_score(self, subject, score):
        self._scores.append(score)
    def avg(self):
        if len(self._scores) == 0:
            return 0
        return sum(self._scores) / len(self._scores)
    def rank(self):
        a = self.avg()
        if a >= 8:
            return "Excellent"
        elif a >= 6.5:
            return "Good"
        elif a >= 5:
            return "Average"
        else:
            return "Poor"
    def __str__(self):
        return self._name + " " + f"{self.avg():.2f}" + " " + self.rank()
name = input()
n = int(input())
stu = Student(name)
for i in range(n):
    subject, score = input().split()
    score = float(score)
    stu.add_score(subject, score)
print(stu)
```

```python
class Book:
    def __init__(self, title, author, year):
        self._title = title
        self._author = author
        self._year = year
class Library:
    def __init__(self):
        self._books = []
    def add(self, book):
        self._books.append(book)
    def count_by_author(self, name):
        cnt = 0
        for b in self._books:
            if b._author == name:
                cnt += 1
        return cnt
    def find_by_year(self, y):
        cnt = 0
        for b in self._books:
            if b._year == y:
                cnt += 1
        return cnt
lib = Library()
n = int(input())
for i in range(n):
    line = input()
    if line.startswith("ADD"):
        tmp = line[4:].split(";")
        title = tmp[0]
        author = tmp[1]
        year = int(tmp[2])
        lib.add(Book(title, author, year))
    elif line.startswith("COUNTYEAR"):
        y = int(line.split()[1])
        print(lib.find_by_year(y))
    elif line.startswith("COUNT"):
        author = line.split()[1]
        print(lib.count_by_author(author))
```

```python
from math import *
def vector(a, b):
    return [b[0] - a[0], b[1] - a[1], b[2] - a[2]]
def cross(u, v):
    return [u[1] * v[2] - u[2] * v[1], u[2] * v[0] - u[0] * v[2], u[0] * v[1] - u[1] * v[0]]
def dot(u, v):
    return u[0] * v[0] + u[1] * v[1] + u[2] * v[2]
def norm(u):
    return sqrt(dot(u, u))
A = list(map(float, input().split()))
B = list(map(float, input().split()))
C = list(map(float, input().split()))
D = list(map(float, input().split()))
AB = vector(A, B)
AC = vector(A, C)
BC = vector(B, C)
BD = vector(B, D)
n1 = cross(AB, AC)
n2 = cross(BC, BD)
cos_phi = abs(dot(n1, n2)) / (norm(n1) * norm(n2))
phi = degrees(acos(cos_phi))
print(f"{phi:.2f}")
```

```python
class ComplexNumber:
    def __init__(self, real, imaginary):
        self._real = real
        self._imag = imaginary
    def multiply(self, other):
        real = self._real * other._real - self._imag * other._imag
        imag = self._real * other._imag + self._imag * other._real
        return ComplexNumber(real, imag)
a, b = map(float, input().split())
c, d = map(float, input().split())
z1 = ComplexNumber(a, b)
z2 = ComplexNumber(c, d)
res = z1.multiply(z2)
print(res._real, res._imag)
```

```python
from math import *
class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self._a = edge1
        self._b = edge2
        self._c = edge3
    def getArea(self):
        a = self._a
        b = self._b
        c = self._c
        if a + b <= c or a + c <= b or b + c <= a:
            return -1
        p = (a + b + c) / 2
        return sqrt(p * (p - a) * (p - b) * (p - c))
a, b, c = map(int, input().split())
tri = Triangle(a, b, c)
area = tri.getArea()
if area == -1:
    print("invalid")
else:
    print(f"{area:.2f}")
```

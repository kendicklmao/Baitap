```python
def merge(a, b):
    i = 0
    j = 0
    n = len(a)
    m = len(b)
    ans = []
    while (i < n and j < m):
        if (a[i] < b[j]):
            ans.append(a[i])
            i += 1
        else:
            ans.append(b[j])
            j += 1
    if (i == n):
        for k in range(j, m):
            ans.append(b[k])
    else:
        for k in range(i, n):
            ans.append(a[k])
    return ans
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(*merge(a, b))
```

```python
n = input()
s = list(n)
l = len(s)
ans = ""
for i in range(0, l):
    Min = min(s[i + 1:l])
    if (Min != s[i]):
        for j in range(l - 1, i, -1):
            if (s[j] == Min):
                s[i], s[j] = s[j], s[i]
                break
        break
if (s[0] == "0"):
    for i in range(1, l):
        ans += s[i]
else:
    for i in range(0, l):
        ans += s[i]
print(ans)
```

```python
def solution(a, b):
    j = 0
    i = 0
    n = len(a)
    m = len(b)
    while (i < n and j < m):
        if (a[i].upper() == b[j]):
            i += 1
            j += 1
        else:
            if (a[i] == a[i].upper()):
                return False
            else:
                i += 1
    if (j == m):
        return True
    return False
t = int(input())
for i in range(0, t):
    a = input()
    b = input()
    if (solution(a, b) == True):
        print("YES")
    else:
        print("NO")
```

```python
def nextPermutation(nums):
    i = len(nums) - 1
    n = i
    while i > 0 and nums[i] <= nums[i - 1]:
        i -= 1
    if i == 0:
        nums.reverse()
        return
    j = n
    while j > i and nums[j] <= nums[i - 1]:
        j -= 1
    nums[j], nums[i - 1] = nums[i - 1], nums[j]
    nums[i:] = reversed(nums[i:])
n = int(input())
nums = list(map(int, input().split()))
nextPermutation(nums)
print(*nums)
```

```python
def solution(houses, heaters):
    len_houses = len(houses)
    len_heaters = len(heaters)
    houses.sort()
    heaters.sort()
    i = 0
    min_radius = 0
    for house in houses:
        while (i + 1 < len_heaters and abs(heaters[i + 1] - house) <= abs(heaters[i] - house)):
            i += 1
        min_radius = max(min_radius, abs(heaters[i] - house))
    return min_radius
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(solution(houses, heaters))
```

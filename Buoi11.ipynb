```python
try:
    a = int(input())
    visited = set()
    while True:
        Sum = 0
        while (a != 0):
            Sum += (a % 10) * (a % 10)
            a //= 10
        if (Sum in visited):
            print("NO")
            break
        visited.add(Sum)
        if (Sum == 1):
            print("YES")
            break
        a = Sum
except ValueError:
    print("Value error")
```

```python
t = int(input())
check = True
while (t != 0):
    regex = input()
    for i in range(0, len(regex)):
        if (regex[i] == "+"):
            if (i == 0):
                print("False")
                check = False
                break
            elif (regex[i - 1] == "+" or regex[i - 1] == "*"):
                print("False")
                check = False
                break
        elif (regex[i] == "*"):
            if (i == 0):
                print("False")
                check = False
                break
            elif (regex[i - 1] == "+" or regex[i - 1] == "*"):
                print("False")
                check = False
                break
    if (check == True):
        print("True")
    t -= 1
```

```python
from math import isqrt
def prime(n):
    if (n < 2):
        return False
    for i in range(2, isqrt(n) + 1):
        if (n % i == 0):
            return False
    return True
n = int(input())
print(prime(n))
```

```python
from math import isqrt
def perfect(n):
    Sum = 0
    for i in range(1, isqrt(n) + 1):
        if (n % i == 0):
            Sum += i
            if (i != (n // i)):
                Sum += (n // i)
    return (Sum - n) == n
n = int(input())
print(perfect(n))
```

```python
inp = list(map(str, input().split()))
if (inp[1] == "+"):
    print(f"{float(inp[0]) + float(inp[2]):.2f}")
elif (inp[1] == "-"):
    print(f"{float(inp[0]) - float(inp[2]):.2f}")
elif (inp[1] == "*"):
    print(f"{float(inp[0]) * float(inp[2]):.2f}")
else:
    print(f"{float(inp[0]) / float(inp[2]):.2f}")
```

```python
x, y = map(int, input().split())
x1, y1 = bin(x)[2::], bin(y)[2::]
dis = 0
if (len(x1) > len(y1)):
    Len = len(x1)
    y1 = "0" * (len(x1) - len(y1)) + y1
else:
    Len = len(y1)
    x1 = "0" * (len(y1) - len(x1)) + x1
for i in range(0, Len):
    if (x1[i] != y1[i]):
        dis += 1
print(dis)
```

```python
n = int(input())
Sum = 0
while (n != 0):
    Sum += (n % 10)
    n //= 10
print(Sum)
```

```python
def linear(a, b):
    dick = {}
    Len = len(a)
    for i in range(0, Len):
        if (a[i] not in dick):
            dick[a[i]] = b[i]
        else:
            if (dick[a[i]] != b[i]):
                return False
    return True
a, b = map(str, input().split())
print(linear(a, b) and linear(b, a))
```

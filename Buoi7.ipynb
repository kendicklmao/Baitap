```python
def binary_search(nums, left, right, target):
    while (left <= right):
        mid = (left + right) // 2
        if (nums[mid] == target):
            return mid
        elif (nums[mid] > target):
            right = mid - 1
        else:
            left = mid + 1
    
    return -1
nums = list(map(int, input().split()))
target = int(input())
n = len(nums)
print(binary_search(nums, 0, n - 1, target))
```

```python
def bubble_sort(nums, n):
    for i in range(0, n):
        check = False
        for j in range(0, n - i - 1):
            if (nums[j] > nums[j + 1]):
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
                check = True
        if (check == False):
            break
    return

def selection_sort(nums, n):
    for i in range(0, n - 1):
        Min = i
        for j in range(i + 1, n):
            if (nums[j] < nums[Min]):
                Min = j
        nums[i], nums[Min] = nums[Min], nums[i]
    return
nums = list(map(int, input().split()))
n = len(nums)
selection_sort(nums, n)
print(*nums)
```

```python
def occur_most(nums):
    Max = 0
    cnt = {}
    for i in nums:
        if (i not in cnt):
            cnt[i] = 1
        else:
            cnt[i] += 1
    for k, v in cnt.items():
        if (v > Max):
            Max = v
            ans = k
    return ans
nums = list(map(int, input().split()))
print(occur_most(nums))
```

```python
def interval(intervals, queries):
    ans = []
    for i in queries:
        Min = 100000000000
        for left, right in intervals:
            if left <= i <= right:
                Min = min(Min, right - left + 1)
        if Min == 100000000000:
            ans.append(-1)
        else:
            ans.append(Min)
    return ans
nums = list(map(int, input().split()))
queries = list(map(int, input().split()))
intervals = []
tmp = []
for i in range(0, len(nums)):
    tmp.append(nums[i])
    if (i % 2 == 1):
        intervals.append(tmp)
        tmp = []
print(interval(intervals, queries))
```

```python
nums = list(map(int, input().split()))
target = int(input())
prefix = [0]
n = len(nums)
Max = 0
for i in range(1, n):
    prefix.append((prefix[i - 1] + nums[i - 1]))
    Max = max(Max, (prefix[i - 1] + nums[i - 1]) % target)
prefix.append(prefix[n - 1] + nums[n - 1])
if (Max == target - 1):
    print(Max)
else:
    for i in range(1, n):
        for j in range(0, i):
            if ((prefix[i] - prefix[j]) % target > Max):
                Max = (prefix[i] - prefix[j]) % target
    print(Max)
```

```python
s = input().strip()
ans = {}
if s:
    pairs = s.split()
    for pair in pairs:
        key, value = pair.split(':')
        if key not in ans:
            ans[key] = []
        ans[key].append(value)
print(ans)
```

```python
nums = list(map(int, input().strip().split()))
ans = {
    'positives': 0,
    'negatives': 0,
    'zeros': 0
}
for i in nums:
    if (i > 0):
        ans['positives'] += 1
    elif (i < 0):
        ans['negatives'] += 1
    else:
        ans['zeros'] += 1
print(ans)
```

```python
d1 = input()
d2 = input()
ans = {}
for i in range(2, len(d1), 8):
    ans[d1[i]] = d1[i + 4]
for i in range(2, len(d2), 8):
    if d2[i] in ans:
        ans[d2[i]] = int(ans[d2[i]]) + int(d2[i + 4])
    else:
        ans[d2[i]] = d2[i + 4]
ans = dict(sorted(ans.items()))
print(ans)
```

```python
nums = list(map(int, input().strip().split()))
k = int(input().strip())
n = len(nums)
ans = []
for i in range(0, n):
    for j in range(i + 1, n):
        if nums[i] + nums[j] == k:
            ans.append((nums[i], nums[j]))
ans.sort(key=lambda x: x[0])
print(ans)
```

```python
nums = list(map(int, input().split()))
cnt = {}
for i in nums:
    if i in cnt:
        cnt[i] += 1
    else:
        cnt[i] = 1
Max = 0
for v in cnt.values():
    if v > Max:
        Max = v
ans = None
for k in cnt:
    if cnt[k] == Max:
        if ans is None or k < ans:
            ans = k
print(ans)
```

```python
data = input().split()
dic = {}
i = 0
while i < len(data):
    key = data[i]
    value = data[i + 1]
    dic[key] = value
    i += 2
new_dic = {}
for k in dic:
    v = dic[k]
    new_dic[v] = k
print(new_dic)
```

```python
data = input().split()
k = int(input())
dic = {}
i = 0
while i < len(data):
    key = data[i]
    value = int(data[i + 1])
    dic[key] = value
    i += 2
new_dic = {}
for key in dic:
    if dic[key] > k:
        new_dic[key] = dic[key]
print(new_dic)
```

```python
words = input().split()
dic = {}
for i in range(len(words)):
    w = words[i]
    if w in dic:
        dic[w].append(i)
    else:
        dic[w] = [i]
ans = {}
for k in dic:
    ans[k] = tuple(dic[k])
print(ans)
```
